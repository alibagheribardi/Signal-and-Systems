$$
T_{sin}(x) = \sum_{k=1}^5 \sin(2k\pi x) ~~~, ~~~ T_{cos}(x) = \sum_{k=1}^5 \cos k\pi x)
$$



```python
import numpy as np
import matplotlib.pyplot as plt

# Define the x values (5000 points between -10 and 10)
x_vals = np.linspace(-10, 10, 5000)

# Initialize arrays to store the sums of sine and cosine terms
T_sin = np.zeros_like(x_vals)  # For sine terms
T_cos = np.zeros_like(x_vals)  # For cosine terms

# Define the range of k values (1 to 5)
k_values = list(range(1, 6))

# Create a figure with subplots (5 rows, 2 columns)
fig, ax = plt.subplots(len(k_values), 3, figsize=(10, 8), sharex=True)

# Loop over each k value and compute the sums
for idx, k in enumerate(k_values):
    # Add the sine term to T_sin
    T_sin += np.sin(2 * np.pi * k * x_vals)
    # Add the cosine term to T_cos
    T_cos += np.cos(2 * np.pi * k * x_vals)

    # Plot the current sum of sine terms in the first column
    ax[idx, 0].plot(x_vals, T_sin, label=f"Sum up to k={k}", color=f"C{idx}")
    ax[idx, 0].set_ylabel("T_sin")
    ax[idx, 0].legend()
    ax[idx, 0].grid()
    ax[idx, 0].set_xlim(-2, 2)  # Set x-axis limits
    ax[idx, 0].set_ylim(-3, 3)  # Set y-axis limits

    # Plot the current sum of cosine terms in the second column
    ax[idx, 1].plot(x_vals, T_cos, label=f"Sum up to k={k}", color=f"C{idx}")
    ax[idx, 1].set_ylabel("T_cos")
    ax[idx, 1].legend()
    ax[idx, 1].grid()
    ax[idx, 1].set_xlim(-2, 2)  # Set x-axis limits
    ax[idx, 1].set_ylim(-3, 3)  # Set y-axis limits

    # Plot the current sum of cosine terms in the second column
    ax[idx, 2].plot(x_vals, T_cos + T_sin, label=f"Sum up to k={k}", color=f"C{idx}")
    ax[idx, 2].set_ylabel("T_cos")
    ax[idx, 2].legend()
    ax[idx, 2].grid()
    ax[idx, 2].set_xlim(-3, 3)  # Set x-axis limits
    ax[idx, 2].set_ylim(-4, 4)  # Set y-axis limits


# Set x-axis label for the last row
ax[-1, 0].set_xlabel("x")
ax[-1, 1].set_xlabel("x")

# Adjust the layout to prevent overlap
plt.tight_layout()

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the Fourier Series terms for a square wave
def fourier_series_square_wave(x, N):
    result = np.zeros_like(x)
    for n in range(1, N+1, 2):  # Only odd terms (1, 3, 5, ...)
        result += (4 / np.pi) * (1 / n) * np.sin(n * x)  # Fourier coefficient for square wave
    return result

# Define the x values
x_vals = np.linspace(-2 * np.pi, 2 * np.pi, 1000)

# Create the Fourier series approximation
N_values = [1, 3, 5, 7, 9]  # Increasing number of terms in the Fourier series
colors = ['r', 'g', 'b', 'c', 'm']

# Plotting the Fourier series approximations
plt.figure(figsize=(12, 6))

# Plot the Fourier series with increasing number of terms
for idx, N in enumerate(N_values):
    approx = fourier_series_square_wave(x_vals, N)
    plt.plot(x_vals, approx, label=f"Fourier Series with {N} terms", color=colors[idx])

# Final plot settings
plt.title(r"Fourier Series Approximation of the Square Wave: $f(x) = \frac{4}{\pi} \sum_{n=1,3,5,\dots} \frac{1}{n} \sin(nx)$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.legend()
plt.grid(True)
plt.show()
```

## The Fourier Hypothesis
The Fourier Hypothesis asserts that any periodic function $f(x)$ with period $T$ can be written as a sum of sinusoids (sine and cosine functions) of integer multiples of a fundamental frequency. Specifically, for a periodic function $f(x)$, we can express it in the following form:

$$
f(x) = a_0 + \sum_{n=1}^{\infty} \left( a_n \cos(n\omega x) + b_n \sin(n\omega x) \right),
$$

where:

$a_0$ is the mean or average value of the function over one period, which represents the DC (direct current) component of the signal.

$a_n$ and $b_n$ are the Fourier coefficients, which indicate the amplitude of the cosine and sine components, respectively, corresponding to the $n$-th harmonic (frequency).

$\omega = \frac{2\pi}{T}$ is the fundamental angular frequency associated with the periodic function $f(x)$, which is the frequency of the first harmonic.

The summation includes all harmonics $n = 1, 2, 3, \dots$, with each harmonic corresponding to a multiple of the fundamental frequency.

This decomposition is particularly valuable because the sine and cosine functions are well-understood, orthogonal functions, making them an ideal basis for constructing more complex periodic waveforms!

## Mathematical Representation
The Fourier series for a periodic function $f(x)$ with period $T$ is given by:

$$
f(x) = a_0 + \sum_{n=1}^{\infty} \left[ a_n \cos(n\omega x) + b_n \sin(n\omega x) \right],
$$

where the Fourier coefficients $a_0$, $a_n$, and $b_n$ are calculated using integrals over one period $T$. These coefficients are given by the following formulas:

$$
a_0 = \frac{1}{T} \int_0^T f(x) \, dx,
$$

$$
a_n = \frac{2}{T} \int_0^T f(x) \cos(n\omega x) \, dx,
$$

$$
b_n = \frac{2}{T} \int_0^T f(x) \sin(n\omega x) \, dx.
$$



```python
import numpy as np
import matplotlib.pyplot as plt

# Define the original function
def f(x):
    return -x**2 + 1

# Define the periodic extension
def periodic_f(x):
    # Use modulo to create periodicity
    x_mod = x % 2
    # Shift the values to the range [-1, 1]
    x_shifted = np.where(x_mod > 1, 2 - x_mod, x_mod)
    return f(x_shifted)

# Generate x values
x_values = np.linspace(-5, 5, 1000)  # Change the range as needed
y_values = periodic_f(x_values)

# Plot the function
plt.figure(figsize=(10, 4))
plt.plot(x_values, y_values, label='Periodic extension of f(x) = -x^2 + 1')
plt.title('Periodic Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(0, color='black',linewidth=0.5, ls='--')
plt.axvline(0, color='black',linewidth=0.5, ls='--')
plt.ylim(-2,5)
plt.grid()
plt.legend()
plt.show()
```

\Large The Fourier coefficients for the function $f(x) = -x^2 + 1$ defined on the interval $[-1, 1]$ are given by:

1. Zeroth Coefficient:
   $$
   a_0 = \frac{1}{2} \int_{-1}^{1} (-x^2 + 1) \, dx = \frac{2}{3}
   $$

2. Cosine Coefficients:
   $$
   a_n = \int_{-1}^{1} (-x^2 + 1) \cos(n \pi x) \, dx
   $$
   Since the integral of the cosine term evaluates to zero for $n \geq 1$, we have:
   $$
   a_n = -\int_{-1}^{1} x^2 \cos(n \pi x) \, dx = \frac{4(-1)^{n+1}}{(n \pi)^2} \quad (n \geq 1)
   $$

3. **Sine Coefficients**:
   $$
   b_n = \int_{-1}^{1} (-x^2 + 1) \sin(n \pi x) \, dx = 0 \quad (n \geq 1)
   $$

In summary, the coefficients are:
- Zeroth Coefficient:
  $$
  a_0 = \frac{2}{3}
  $$

- Cosine Coefficients:
  $$
  a_n = \frac{4(-1)^{n+1}}{(n \pi)^2} \quad (n \geq 1)
  $$

- Sine Coefficients:
  $$
  b_n = 0 \quad (n \geq 1)
  $$

Thus, the Fourier series representation of $f(x)$ becomes:

$$
f(x) = \frac{2}{3} + \sum_{n=1}^{\infty} \left( \frac{4(-1)^{n+1}}{(n \pi)^2} \cos(n \pi x) + 0 \cdot \sin(n \pi x) \right)
$$

This simplifies to:

$$
f(x) = \frac{2}{3} + \sum_{n=1}^{\infty} \frac{4(-1)^{n+1}}{(n \pi)^2} \cos(n \pi x)
$$

```python
import numpy as np
#import matplotlib.pyplot as plt

# Define the original function
def f(x):

    y= -x**2 + 1
    return y

def fourier_series(x, N):
    a0 = 2 / 3  # The DC term
    series_sum = a0
    for n in range(1, N + 1):
        an = 4 * (-1)**(n + 1) / (n * np.pi)**2
        series_sum += an * np.cos(n * np.pi * x)
    return series_sum
#print(f(0.8) , fourier_series(0.8,5) )

x_values = np.linspace(-1, 1, 1000)
y_values = f(x_values)

N = int(input("Enter the number of terms (N) for the Fourier series approximation: "))

fourier_values = fourier_series(x_values, N)

# Calculate the error between the original function and the Fourier series approximation
error = np.abs(y_values - fourier_values)
max_error = np.max(error)
```

```python

import matplotlib.pyplot as plt


# Create a figure and a set of subplots
_ , axs = plt.subplots(1, 2)

# Plot original function vs Fourier approximation
axs[0].plot(x_values, y_values, label='$f(x) = -x^2 + 1$', color='blue', lw=1, alpha=0.6, ls='--')
axs[0].plot(x_values, fourier_values, label=f'Fourier Approximation (N={N})', color='red', linestyle='-', lw=.5)
axs[0].set_title('Fourier Series Approximation of $f(x) = -x^2 + 1$')
axs[0].set_xlabel('x')
axs[0].set_ylabel('f(x)')
axs[0].axhline(0, color='black', linewidth=3)
axs[0].axvline(0, color='black', linewidth=3)
axs[0].set_ylim(-1, 2)
axs[0].set_xlim(-1, 1)
axs[0].grid()
axs[0].legend()

# Plot the error
axs[1].plot(x_values, error, linestyle='--', color='purple', label=f'Error (N={N})')
axs[1].set_title('Error between Original Function and Fourier Approximation')
axs[1].set_xlabel('x')
axs[1].set_ylabel('Error')
axs[1].axhline(0, color='black', linewidth=0.5)
axs[1].set_ylim(0, 0.1)
axs[1].set_xlim(-1, 1)
axs[1].grid()
axs[1].legend()

# Display maximum error in the error plot
axs[1].annotate(f'Max Error: {max_error:.5f}', xy=(0.05, 0.08), fontsize=12, color='black')

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```

## Complex Periodic Functions

For a periodic function $ f(t) $ with period $ T $, the Fourier series representation can be expressed in terms of complex exponentials as follows:

$$
f(t) = \sum_{n=-\infty}^{\infty} c_n e^{i \frac{2\pi n}{T} t}
$$

where $ c_n $ are the Fourier coefficients given by:

$$
c_n = \frac{1}{T} \int_{0}^{T} f(t) e^{-i \frac{2\pi n}{T} t} \, dt
$$

-  Period $ T $: The length of one complete cycle of the periodic function.
- Complex Exponentials: The term $ e^{i \frac{2\pi n}{T} t} $ can be expressed using Euler's formula:
  $$
  e^{i\theta} = \cos(\theta) + i\sin(\theta)
  $$
  This means that each term in the series can be thought of as a combination of sine and cosine functions.
    
- Fourier Coefficients $ c_n $: These coefficients determine the amplitude each frequency component in the series. They are calculated by integrating the product of the function and the complex exponential over one period.

- Orthogonality: The complex exponentials $ e^{i \frac{2\pi n}{T} t} $ are orthogonal over the interval $ [0, T] $.

- Reconstruction: The original function can be reconstructed from its Fourier series by summing an infinite number of terms, which allows for the analysis of the frequency content of the function.




## Coefficients for Real Functions

For a real-valued function $ f(t) $, the coefficients $ c_n $ must satisfy the following relationship:

$$
c_{-n} = \overline{c_n}
$$

This means that the coefficient for the negative frequency is the complex conjugate of the coefficient for the positive frequency.

We can express the coefficients $ c_n $ in terms of their real and imaginary parts:

$$
c_n = a_n - i b_n
$$

where $ a_n $ and $ b_n $ are real-valued functions. Thus, we can write:

$$
f(t) = \sum_{n=0}^{\infty} \left( a_n \cos\left(\frac{2\pi n}{T} t\right) + b_n \sin\left(\frac{2\pi n}{T} t\right) \right)
$$



## EXERCISE.

- Even Function: If $ f(t) $ is an even function (i.e., $ f(t) = f(-t) $), then all sine terms vanish because sine is an odd function. Thus, the Fourier series contains only cosine terms:

   $$
   f(t) = \sum_{n=0}^{\infty} a_n \cos\left(\frac{2\pi n}{T} t\right)
   $$
- Odd Function: If $ f(t) $ is an odd function (i.e., $ f(t) = -f(-t) $), then all cosine terms vanish because cosine is an even function. Thus, the Fourier series contains only sine terms:

   $$
   f(t) = \sum_{n=1}^{\infty} b_n \sin\left(\frac{2\pi n}{T} t\right)
   $$

For any real-valued periodic function $ f(t) $:

- If $ f(t) $ is even, its Fourier series consists only of cosine terms.
- If $ f(t) $ is odd, its Fourier series consists only of sine terms.
