```python
import numpy as np
import matplotlib.pyplot as plt

# Generate values for t from 0 to 2Ï€
t = np.linspace(0, 2 * np.pi, 100)

# Calculate cos^2(t), 1/2 + cos(2t), and cos(t)
y = np.cos(t)**2
z = 1/2*np.cos(2*t)  # Corrected to match your original intention
w = np.cos(t)

# Create the plot
plt.figure(figsize=(10, 4))
plt.plot(t, y, lw=2, color='black', label=r'$\cos^2(t)$')
plt.plot(t, z, lw=2, color='blue', label=r'$\frac{1}{2}\cos(2t)$')
plt.plot(t, w, lw=2, color='red', label=r'$\cos(t)$' , ls= '--')
plt.title(r'Plot of $\cos^2(t)$, $\frac{1}{2} + \cos(2t)$, and $\cos(t)$ from $0$ to $2\pi$')
plt.xlabel('t (radians)')
plt.ylabel('Function Value')
plt.xticks(np.arange(0, 2 * np.pi + 0.1, np.pi / 2),
           [r'$0$', r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$'])
plt.yticks(np.arange(-1, 2, 0.5))
plt.grid()
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate values for t from -Ï€ to Ï€
t = np.linspace(-np.pi, np.pi, 1000)

# Define the function f(t)
f = np.where((t > -np.pi/2) & (t < np.pi/2), 4, 0)

# Create the plot
plt.figure(figsize=(10, 4))
plt.plot(t, f, lw=2, color='black', label=r'$f(t)$')
plt.title(r'Plot of the function $f(t)$ on $[-\pi, \pi]$')
plt.xlabel('t (radians)')
plt.ylabel(r'$f(t)$')
plt.xticks(np.arange(-np.pi, np.pi + 0.1, np.pi/2),
           [r'$-\pi$', r'$-\frac{\pi}{2}$', r'$0$', r'$\frac{\pi}{2}$', r'$\pi$'])
plt.yticks(np.arange(0, 5, 1))
plt.grid()
plt.legend()
plt.ylim(-1, 5)  # Set y-limits for better visibility
plt.show()
```

## On $[-\pi , \pi]$ define  $$ f(t) =
\begin{cases}
4 & \text{if } -\frac{\pi}{2} < t < \frac{\pi}{2} \\
0 & \text{otherwise}
\end{cases}  $$
## We have that
$$ f(t) = 2 + 8/\pi( \cos t -1/3cos3t + 1/5cos5t -1/7cos7t + ... ) $$

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate values for t from -Ï€ to Ï€
t = np.linspace(-np.pi, np.pi, 1000)

# Define each term
term1 = 2
term2 = (8 / np.pi) * np.cos(t) + term1
term3 = -(8 / (3 * np.pi)) * np.cos(3 * t) + term2
term4 = (8 / (5 * np.pi)) * np.cos(5 * t) + term3
term5 = -(8 / (7 * np.pi)) * np.cos(7 * t) + term4

f = np.where((t > -np.pi/2) & (t < np.pi/2), 4, 0)

# Create the plot
plt.figure(figsize=(12, 5))

# Plot each term
plt.plot(t, term1 * np.ones_like(t), lw=2, color='gray', label=r'$a_0 = 2$')
plt.plot(t, term2, lw=2, color='blue', label=r'$a_0 + \frac{8}{\pi} \cos(t)$')
plt.plot(t, term3, lw=2, color='orange', label=r'$a_0 + \frac{8}{\pi} \cos(t) - \frac{8}{3\pi} \cos(3t)$')
plt.plot(t, term4, lw=2, color='green', label=r'$a_0 + \frac{8}{\pi} \cos(t) - \frac{8}{3\pi} \cos(3t) + \frac{8}{5\pi} \cos(5t)$')
plt.plot(t, term5, lw=2, color='red', label=r'$a_0 + \frac{8}{\pi} \cos(t) - \frac{8}{3\pi} \cos(3t) + \frac{8}{5\pi} \cos(5t) - \frac{8}{7\pi} \cos(7t)$')
plt.plot(t, f, lw=2, color='black', label=r'$f(t)$', linestyle='--')

# Add titles and labels
plt.title(r'Plot of each term of the function $f(t)$ on $[-\pi, \pi]$')
plt.xlabel('t (radians)')
plt.ylabel(r'$f(t)$')
plt.xticks(np.arange(-np.pi, np.pi + 0.1, np.pi/2),
           [r'$-\pi$', r'$-\frac{\pi}{2}$', r'$0$', r'$\frac{\pi}{2}$', r'$\pi$'])
plt.yticks(np.arange(-5, 5, 1))
plt.grid()
plt.legend()
plt.ylim(-5, 5)  # Set y-limits for better visibility
plt.show()
```
